```python
allVisitedCoords = []
def navigateMap(currentCoords, pathSoFar, shortestCompletedPath, endLocation, theMap):
    global allVisitedCoords
    pathSoFar.append(currentCoords)
    allVisitedCoords.append(currentCoords)

    #stop navigation if this path is longer than one that has already been found
    if shortestCompletedPath != None and len(pathSoFar) >= len(shortestCompletedPath):
        return shortestCompletedPath
    elif currentCoords == endLocation: #we found a new shortest path!
        print("found a new shortest path!!")
        shortestCompletedPath = pathSoFar
        return shortestCompletedPath

    coordsToNorth = (currentCoords[0]-1, currentCoords[1])
    coordsToSouth = (currentCoords[0]+1, currentCoords[1])
    coordsToEast = (currentCoords[0], currentCoords[1]+1)
    coordsToWest = (currentCoords[0], currentCoords[1]-1)

    def areCoordsValid(currentCoords, testingCoords, pathSoFar, theMap) -> bool:
        global allVisitedCoords
        #check if we're moving out of bounds
        if (testingCoords[0] < 0) or (testingCoords[0] > len(theMap)-1): #too far vertical
            return False
        if (testingCoords[1] < 0) or (testingCoords[1] > len(theMap[testingCoords[0]])-1): #too far horizontal
            return False

        if list(pathSoFar).count(testingCoords) > 0: #we've already been here, don't go again
            return False
        currentElevation = theMap[currentCoords[0]][currentCoords[1]]
        testingElevation = theMap[testingCoords[0]][testingCoords[1]]

        if testingElevation > currentElevation+1: #this is too high to climb
            return False
        
        return True
    
    if areCoordsValid(currentCoords, coordsToEast, pathSoFar, theMap):
        shortestCompletedPath = navigateMap(coordsToEast, pathSoFar.copy(), shortestCompletedPath, endLocation, theMap)
    if areCoordsValid(currentCoords, coordsToNorth, pathSoFar, theMap):
        shortestCompletedPath = navigateMap(coordsToNorth, pathSoFar.copy(), shortestCompletedPath, endLocation, theMap)
    if areCoordsValid(currentCoords, coordsToSouth, pathSoFar, theMap):
        shortestCompletedPath = navigateMap(coordsToSouth, pathSoFar.copy(), shortestCompletedPath, endLocation, theMap)
    if areCoordsValid(currentCoords, coordsToWest, pathSoFar, theMap):
        shortestCompletedPath = navigateMap(coordsToWest, pathSoFar.copy(), shortestCompletedPath, endLocation, theMap)

    # print("hit dead end")
    return shortestCompletedPath

shortestPath = navigateMap(startLocation, [], None, endLocation, theMap)
print(shortestPath)
print(len(shortestPath))
```
