```python
scoreTrees = 0
mapOfVisibleTrees = ""
for i, row in enumerate(filteredInput):
    for j, treeHeight in enumerate(row):
        if (i == 0) or (i == len(filteredInput)-1) or (j == 0) or (j == len(row)-1): #mark edge trees as visible
            visibleTrees += 1
            mapOfVisibleTrees += treeHeight
            pass
        else:
            visibleLeft = True
            visibleRight = True
            visibleTop = True
            visibleBottom = True

            for k, otherTreeHeight in enumerate(row[0:j]):
                # print(f"h: {otherTreeHeight}")
                if otherTreeHeight >= treeHeight:
                    visibleLeft = False
                    # print("taller on left: " + otherTreeHeight)
                    break
            for k, otherTreeHeight in enumerate(row[j+1:len(row)]):
                # print(f"h: {otherTreeHeight}")
                if otherTreeHeight >= treeHeight:
                    visibleRight = False
                    # print("taller on right: " + otherTreeHeight)
                    break
            
            for l, rowAgain in enumerate(filteredInput[0:i]):
                # print(f"v: {rowAgain[j]}")
                if rowAgain[j] >= treeHeight:
                    visibleTop = False
                    # print("taller on top: " + rowAgain[j])
                    break
            for l, rowAgain in enumerate(filteredInput[i+1:len(row)]):
                # print(f"v: {rowAgain[j]}")
                if rowAgain[j] >= treeHeight:
                    visibleBottom = False
                    # print("taller on top: " + rowAgain[j])
                    break

            # if visible:
            #     print(f"visible tree at {i}, {j}: {treeHeight}")
            #     visibleTrees += 1
            # else:
            #     print(f"{i}, {j}: {treeHeight}")
            if visibleLeft or visibleRight or visibleTop or visibleBottom:
                visibleTrees += 1
                mapOfVisibleTrees += treeHeight
                # print(f'Visible tree: h={treeHeight}, row={i}, column={j}')
            else:
                mapOfVisibleTrees += "-"
    mapOfVisibleTrees += "\n"
            
print(mapOfVisibleTrees)
visibleTrees
```

```python
highestScenicScore = 0
mapOfScenicScores = ""
mapOfLeftScenicScores = ""
mapOfRightScenicScores = ""
mapOfTopScenicScores = ""
mapOfBottomScenicScores = ""
for i, row in enumerate(filteredInput):
    for j, treeHeight in enumerate(row):
        if (i == 0) or (i == len(filteredInput)-1) or (j == 0) or (j == len(row)-1): #mark edge trees as 0 score
            mapOfScenicScores += "0"
            pass
        else:
            scoreLeft = 0
            scoreRight = 0
            scoreTop = 0
            scoreBottom = 0

            for k, otherTreeHeight in enumerate(reversed(row[0:j])): #check left side
                scoreLeft += 1
                if otherTreeHeight >= treeHeight:
                    break
            for k, otherTreeHeight in enumerate(row[j+1:len(row)]): #check right side
                scoreRight += 1
                if otherTreeHeight >= treeHeight:
                    break
            
            for l, rowAgain in enumerate(reversed(filteredInput[0:i])): #check top side
                scoreTop += 1
                if rowAgain[j] >= treeHeight:
                    break
            for l, rowAgain in enumerate(filteredInput[i+1:len(row)]): #check bottom side
                scoreBottom += 1
                if rowAgain[j] >= treeHeight:
                    break

            scenicScore = scoreLeft * scoreRight * scoreTop * scoreBottom
            if scenicScore > highestScenicScore:
                highestScenicScore = scenicScore
            mapOfScenicScores += " " + str(scenicScore) + " "
            mapOfLeftScenicScores += " " + str(scoreLeft) + " "
            mapOfRightScenicScores += " " + str(scoreRight) + " "
            mapOfTopScenicScores += " " + str(scoreTop) + " "
            mapOfBottomScenicScores += " " + str(scoreBottom) + " "
    mapOfScenicScores += "\n"
    mapOfLeftScenicScores += "\n"
    mapOfRightScenicScores += "\n"
    mapOfTopScenicScores += "\n"
    mapOfBottomScenicScores += "\n"
print("total:")
print(mapOfScenicScores)
print("Left:")
print(mapOfLeftScenicScores)
print("Right:")
print(mapOfRightScenicScores)
print("Top:")
print(mapOfTopScenicScores)
print("Bottom:")
print(mapOfBottomScenicScores)
highestScenicScore
```
