```python
filteredInput = input.splitlines()
filteredInput = [i.split() for i in filteredInput]
filteredInput
```

```python
x = 1
cycles = 0
inProgressTasks = []
i = 0
signalStrengthSum = 0
while i < len(filteredInput)-1:
    cycles += 1
    if cycles == 20 or cycles == 60 or cycles == 100 or cycles == 140 or cycles == 180 or cycles == 220:
        # print(cycles)
        # print(x)
        print(cycles*x)
        signalStrengthSum += cycles*x
    if len(inProgressTasks) == 0:
        if filteredInput[i][0] == "noop":
            # print("noop!")
            i += 1
            continue
        if filteredInput[i][0] == "addx":
            # print("addx detected " + filteredInput[i][1])
            inProgressTasks.append(filteredInput[i][1])
            continue
    else:
        commandToExecute = inProgressTasks.pop()
        x += int(commandToExecute)
        i += 1

signalStrengthSum
```
