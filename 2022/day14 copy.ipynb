```python
#returns true if sand comes to a complete stop within the map.
#returns false if sand fills the hole
def sandPhysics(theMap, location, sandDropPoint, debugOn = False) -> bool:
    def isLocationOccupied(theMap, locationToTest):
        x = locationToTest[0]
        y = locationToTest[1]
        locationContents = theMap[y][x]
        if locationContents == 1 or locationContents == 2:
            return True
        return False
    def isLocationInBounds(theMap, locationToTest):
        x = locationToTest[0]
        y = locationToTest[1]
        if y > len(theMap):
            return False
        if x > len(theMap[0]):
            return False
        if x < 0:
            return False
        return True

    x = location[0]
    y = location[1]
    # print(f"x:{x} y:{y}")

    if not isLocationInBounds(theMap, location):
        print("SAND GOES OUT OF BOUNDS")
        return False

    

    if debugOn:
        theMap[y][x] = 3

    #try to fall straight down
    targetLocation = [x, y+1]
    if not isLocationOccupied(theMap, targetLocation):
        return sandPhysics(theMap, targetLocation, sandDropPoint, debugOn)
    #try to fall down and left
    targetLocation = [x-1, y+1]
    if not isLocationOccupied(theMap, targetLocation):
        return sandPhysics(theMap, targetLocation, sandDropPoint, debugOn)
    #try to fall down and right
    targetLocation = [x+1, y+1]
    if not isLocationOccupied(theMap, targetLocation):
        return sandPhysics(theMap, targetLocation, sandDropPoint, debugOn)
    
    #all possible locations to fall are occupied. mark self on map and come to a rest
    theMap[y][x] = 2
    if location == sandDropPoint:
        return False
    return True
```
