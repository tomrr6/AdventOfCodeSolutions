```python
if len(input) < 4:
    raise ValueError("input must have more than 4 characters")

print(input)

for i, letter in enumerate(input):
    # print(letter)
    if i < 4:
        print("not enough chars yet")
        continue
    test = input[i-4:i]
    print("test: " + test)
    couldBeAnswer = True
    for char in test:
        # print(char)
        if test.count(char) > 1:
            print(char + "found more than once")
            couldBeAnswer = False
            break
    if couldBeAnswer:
        print(i)
        break
```
