```python
class TreeNode:
    def __init__(self, name, parent=None, value=0, children=[]):
        self.name = name
        self.parent = parent
        self.value = value
        self.children = children

    def getName(self):
        return self.name
    def getParent(self):
        return self.parent
    def getValue(self):
        total = 0
        total += self.value
        for child in self.getChildren():
            total += child.getValue()
        return total
    def getChildren(self):
        return self.children

    def addChild(self, newChildTreeNode):
        self.children.append(newChildTreeNode)
```
