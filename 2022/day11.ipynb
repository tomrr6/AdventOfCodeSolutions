```python
filteredInput = input.split("\n\n")
filteredInput = [i.splitlines() for i in filteredInput]
# for i in filteredInput:
#     for j in i:
#         # print(j)
#         # filteredInput[i][j] = filteredInput[i][j].split()
#         j = j.split()
#         print(j)
filteredInput
```

```python
class Monkey:
    def __init__(self, name, items, operationType, operationFactor, testDivisibleBy, targetIfTestTrue, targetIfTestFalse) -> None:
        self.name = name
        self.items = items
        self.operationType = operationType
        self.operationFactor = operationFactor
        self.testDivisibleBy = testDivisibleBy
        self.targetIfTestTrue = targetIfTestTrue
        self.targetIfTestFalse = targetIfTestFalse

    def __str__(self) -> str:
        return self.name        
```

```python
allMonkeys = []
for i, entry in enumerate(filteredInput):
    # print(entry[5].split())
    name = i

    items = []
    for item in entry[1].split():
        if str(item).removesuffix(",").isdigit():
            items.append(item.removesuffix(","))
    
    splitOperationString = entry[2].split()
    operationType = splitOperationString[4]
    operationFactor = splitOperationString[5]

    testDivisibleBy = entry[3].split()[3]

    targetIfTestTrue = entry[4].split()[-1]
    targetIfTestFalse = entry[5].split()[-1]
    
    #append new monkey to list
    allMonkeys.append(Monkey(name, items, operationType, operationFactor, testDivisibleBy, targetIfTestTrue, targetIfTestFalse))

print("initial state:")
for monkey in allMonkeys:
            print(monkey.items)

inspectionCounts = [0]*len(allMonkeys)

for i in range(10000):
    print(f"======================ROUND #{i}======================")
    for j, monkey in enumerate(allMonkeys):
        # print(f"==MONKEY #{j}==")

        itemsCopy = monkey.items.copy()

        while len(itemsCopy) > 0:
            item = itemsCopy.pop(0)

            inspectionCounts[j] += 1

            if monkey.operationFactor == "old":
                factor = int(item)
            else:
                factor = monkey.operationFactor

            if monkey.operationType == "*":
                newWorry = int(item) * int(factor)
            elif monkey.operationType == "+":
                newWorry = int(item) + int(factor)
            else: 
                print("ERROR: Unknown operation type: " + monkey.operationType)

            # divide by 3 because of relief that item wasn't hurt
            # newWorry = int(newWorry/3)

            if int(newWorry) % int(monkey.testDivisibleBy) == 0:
                tossToTarget = monkey.targetIfTestTrue
            else:
                tossToTarget = monkey.targetIfTestFalse

            allMonkeys[int(tossToTarget)].items.append(newWorry)
            # print(f"Item {item} became {newWorry} and was given to Monkey #{tossToTarget}")

        monkey.items = itemsCopy
            
        # for monkey in allMonkeys:
        #     print(monkey.items)

inspectionCounts
```
